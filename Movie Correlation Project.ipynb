```python
#Types of correlation 
#pearson(default) - Measures the linear correlation between two variables.
#kenndal - Measures the ordinal association between two variables.
#spearman - Measures if the relationship between two variables is monotonic.
df.corr(method='pearson')
```

```python
correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix,annot=True)
plt.title('Correlation matrix for numeric features ')

plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
#changing names in to numeric format

df_numerate = df 

for col_name in df_numerate.columns:
    if(df_numerate[col_name].dtype== 'object'):
       df_numerate[col_name] = df_numerate[col_name].astype('category')
       df_numerate[col_name] = df_numerate[col_name].cat.codes
df_numerate.head()
```

```python
df.head()
```

```python
correlation_matrix = df_numerate.corr(method='pearson')

sns.heatmap(correlation_matrix,annot=True)
plt.title('Correlation matrix for numeric features ')

plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```
